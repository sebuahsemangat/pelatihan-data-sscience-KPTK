```python
# menggunakan request dan beutifulshop4
# pip install pandas requests beautifulsoup4

import pandas as pd
import requests
from bs4 import BeautifulSoup

# URL situs web yang akan di-scrape
url = 'https://www.w3schools.com/html/html_tables.asp'

# Mengirim permintaan HTTP untuk mengambil konten halaman
response = requests.get(url)

# Memeriksa apakah permintaan berhasil (status code 200)
if response.status_code != 200:
    print(f"Gagal mengakses halaman: {response.status_code}")
    exit()

# Parsing HTML menggunakan BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Mencari semua elemen tabel di halaman
tables = soup.find_all('table')

# Memastikan ada tabel yang ditemukan
if not tables:
    print("Tidak ada tabel ditemukan di halaman.")
    exit()

# Mengambil tabel pertama (indeks 0)
table = tables[0]

# Menggunakan Pandas untuk membaca tabel dari HTML
# Konversi tabel HTML ke DataFrame
df = pd.read_html(str(table))[0]

# Menampilkan 5 baris pertama dari DataFrame
print("Data dari Web Scraping:")
print(df.head())

# Menyimpan hasil ke file CSV (opsional)
df.to_csv('scraped_data.csv', index=False)
print("\nData telah disimpan ke 'scraped_data.csv'")
```

```python
# Scraping Manual dengan BeautifulSoup

import pandas as pd
import requests
from bs4 import BeautifulSoup

# URL situs web
url = 'https://www.w3schools.com/html/html_tables.asp'

# Mengambil konten halaman
response = requests.get(url)
if response.status_code != 200:
    print(f"Gagal mengakses halaman: {response.status_code}")
    exit()

# Parsing HTML
soup = BeautifulSoup(response.content, 'html.parser')

# Mencari tabel pertama
table = soup.find('table')

# Mengekstrak header tabel
headers = [th.text.strip() for th in table.find_all('th')]

# Mengekstrak baris data
rows = []
for tr in table.find_all('tr')[1:]:  # Lewati baris header
    cells = [td.text.strip() for td in tr.find_all('td')]
    if cells:  # Pastikan baris tidak kosong
        rows.append(cells)

# Membuat DataFrame
df_manual = pd.DataFrame(rows, columns=headers)

# Menampilkan 5 baris pertama
print("Data dari Web Scraping (Manual):")
print(df_manual.head())

# Menyimpan ke CSV
df_manual.to_csv('scraped_data_manual.csv', index=False)
print("\nData telah disimpan ke 'scraped_data_manual.csv'")
```

```python
# sraping instagram
# pip install instaloader pandas
""" Instagram melarang scraping otomatis dalam Terms of Service-nya.
Melakukan scraping dapat menyebabkan akun Anda diblokir atau menghadapi konsekuensi hukum.

gunakan pengambilan data dari profil publik tanpa memerlukan login, untuk menghindari risiko pemblokiran.
"""

import instaloader
import pandas as pd

# Inisialisasi Instaloader
loader = instaloader.Instaloader()

# Tentukan username Instagram (profil publik)
username = 'instagram'  # Ganti dengan username yang diinginkan

try:
    # Mengambil profil
    profile = instaloader.Profile.from_username(loader.context, username)

    # Mengumpulkan data profil
    profile_data = {
        'username': profile.username,
        'followers': profile.followers,
        'following': profile.followees,
        'posts_count': profile.mediacount,
        'bio': profile.biography,
        'is_verified': profile.is_verified
    }

    # Mengumpulkan data posting (batasi ke 5 posting pertama untuk efisiensi)
    posts_data = []
    for post in profile.get_posts():
        posts_data.append({
            'post_url': post.url,
            'likes': post.likes,
            'comments': post.comments,
            'caption': post.caption[:100] if post.caption else '',  # Batasi panjang caption
            'date': post.date
        })
        if len(posts_data) >= 5:  # Hentikan setelah 5 posting
            break

    # Membuat DataFrame untuk data profil
    profile_df = pd.DataFrame([profile_data])

    # Membuat DataFrame untuk data posting
    posts_df = pd.DataFrame(posts_data)

    # Menampilkan hasil
    print("\nData Profil:")
    print(profile_df)
    print("\nData Posting (5 posting pertama):")
    print(posts_df)

    # Menyimpan hasil ke CSV
    profile_df.to_csv('instagram_profile.csv', index=False)
    posts_df.to_csv('instagram_posts.csv', index=False)
    print("\nData telah disimpan ke 'instagram_profile.csv' dan 'instagram_posts.csv'")

except Exception as e:
    print(f"Error: {e}")
```

```python
from ntscraper import Nitter
import pandas as pd
import time

# Inisialisasi scraper dengan logging dan instance Nitter tertentu
scraper = Nitter(log_level=1, skip_instance_check=False)

# Tentukan username dan jumlah tweet
username = 'elonmusk'  # Ganti dengan username valid
tweet_count = 10

try:
    # Mengambil tweet dari akun
    print(f"Mengambil {tweet_count} tweet dari @{username}...")
    tweets_data = scraper.get_tweets(username, mode='user', number=tweet_count)

    # Periksa apakah data tweet kosong
    if not tweets_data['tweets']:
        print(f"Tidak ada tweet ditemukan untuk @{username}. Mungkin akun privat, tidak ada tweet, atau instance Nitter bermasalah.")
        exit()

    # Membuat dictionary untuk menyimpan data
    data = {
        'link': [],
        'text': [],
        'user': [],
        'likes': [],
        'retweets': [],
        'comments': []
    }

    # Mengisi data dari tweet
    for tweet in tweets_data['tweets']:
        data['link'].append(tweet['link'])
        data['text'].append(tweet['text'])
        data['user'].append(tweet['user']['name'])
        data['likes'].append(tweet['stats']['likes'])
        data['retweets'].append(tweet['stats']['retweets'])
        data['comments'].append(tweet['stats']['comments'])

    # Membuat DataFrame
    df = pd.DataFrame(data)

    # Menampilkan 5 baris pertama
    print("\nData Tweet:")
    print(df.head())

    # Menyimpan ke CSV
    df.to_csv('tweets_data.csv', index=False)
    print("\nData telah disimpan ke 'tweets_data.csv'")

except Exception as e:
    print(f"Error: {e}")
    if "Cannot choose from an empty sequence" in str(e):
        print("Kemungkinan penyebab:")
        print("- Username tidak valid atau akun privat.")
        print("- Tidak ada tweet tersedia.")
        print("- Instance Nitter tidak berfungsi. Coba instance lain atau periksa koneksi.")
        print("- Rate limit tercapai. Tunggu beberapa saat.")
    print("Solusi: Periksa username, coba instance Nitter lain, atau gunakan Selenium/API.")

finally:
    # Tambahkan delay untuk menghindari rate limit
    time.sleep(2)
```

```python
# mengambil data dari API

import requests
import json

def fetch_data_from_api(api_url):
    try:
        # Send GET request to the API
        response = requests.get(api_url)

        # Check if request was successful
        response.raise_for_status()

        # Parse JSON response
        data = response.json()

        # Save data to a file
        with open('api_data.json', 'w') as file:
            json.dump(data, file, indent=4)

        print("Data successfully scraped and saved to api_data.json")
        return data

    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred: {http_err}")
    except requests.exceptions.ConnectionError as conn_err:
        print(f"Connection error occurred: {conn_err}")
    except requests.exceptions.Timeout as timeout_err:
        print(f"Timeout error occurred: {timeout_err}")
    except requests.exceptions.RequestException as req_err:
        print(f"An error occurred: {req_err}")
    except ValueError as json_err:
        print(f"JSON decode error: {json_err}")

    return None

def main():
    # Example public API (JSONPlaceholder)
    api_url = "https://jsonplaceholder.typicode.com/posts"
    data = fetch_data_from_api(api_url)

    # Print first few records if data is fetched successfully
    if data:
        print("\nFirst 2 records:")
        for item in data[:2]:
            print(json.dumps(item, indent=4))

if __name__ == "__main__":
    main()
```

```python
from pytrends.request import TrendReq
import pandas as pd
from datetime import datetime

def scrape_google_trends(keyword, timeframe='today 5-y', geo=''):
    try:
        # Initialize pytrends
        pytrends = TrendReq(hl='en-US', tz=360)

        # Build payload
        pytrends.build_payload(kw_list=[keyword], timeframe=timeframe, geo=geo)

        # Get interest over time
        interest_over_time = pytrends.interest_over_time()

        # Get interest by region
        interest_by_region = pytrends.interest_by_region()

        # Save data to CSV
        timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
        interest_over_time.to_csv(f'google_trends_time_{keyword}_{timestamp}.csv')
        interest_by_region.to_csv(f'google_trends_region_{keyword}_{timestamp}.csv')

        print(f"Data successfully scraped and saved for keyword: {keyword}")
        print(f"Time data saved to: google_trends_time_{keyword}_{timestamp}.csv")
        print(f"Region data saved to: google_trends_region_{keyword}_{timestamp}.csv")

        return interest_over_time, interest_by_region

    except Exception as e:
        print(f"An error occurred: {e}")
        return None, None

def main():
    # Example keyword
    keyword = "python programming"

    # Scrape Google Trends data
    time_data, region_data = scrape_google_trends(keyword)

    # Display sample data
    if time_data is not None and not time_data.empty:
        print("\nSample Interest Over Time (first 5 rows):")
        print(time_data.head())

    if region_data is not None and not region_data.empty:
        print("\nSample Interest by Region (first 5 rows):")
        print(region_data.head())

if __name__ == "__main__":
    main()
```

```python
import requests
import pandas as pd
from datetime import datetime
import time
import uuid

def scrape_tokopedia(keyword, max_pages=5, delay=5):
    try:
        # Initialize variables
        url_target = 'https://gql.tokopedia.com/graphql/SearchProductQueryV4'
        headers = {
            'authority': 'gql.tokopedia.com',
            'accept': '*/*',
            'accept-language': 'en-US,en;q=0.9,id;q=0.8',
            'content-type': 'application/json',
            'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36'
        }
        all_products = []
        unique_id = str(uuid.uuid4())

        for page in range(1, max_pages + 1):
            print(f"Scraping page {page} for keyword: {keyword}")
            query = f'[{{"operationName":"SearchProductQueryV4","variables":{{"params":"device=desktop&navsource=home&ob=23&page={page}&q={keyword}&related=true&rows=60&safe_search=false&scheme=https&shipping=&source=universe&srp_component_id=01.02.01.01&st=product&start={(page-1)*60}&topads_bucket=true&unique_id={unique_id}&user_addressId=&user_cityId=176&user_districtId=2274&user_id=&user_lat=&user_long=&user_postCode=&user_warehouseId=12210375&variants="}},"query":"query SearchProductQueryV4($params: String!) {{ ace_search_product_v4(params: $params) {{ header {{ totalData totalDataText processTime responseCode errorMessage additionalParams keywordProcess componentId __typename }} data {{ banner {{ position text imageUrl url componentId trackingOption __typename }} backendFilters isQuerySafe ticker {{ text query typeId componentId trackingOption __typename }} redirection {{ redirectUrl departmentId __typename }} related {{ position trackingOption relatedKeyword otherRelated {{ keyword url product {{ id name price imageUrl rating countReview url priceStr wishlist shop {{ city isOfficial isPowerBadge __typename }} ads {{ adsId: id productClickUrl productWishlistUrl shopClickUrl productViewUrl __typename }} badges {{ title imageUrl show __typename }} __typename }} __typename }} suggestion {{ query url trackingOption componentId __typename }} products {{ id name price priceStr shop {{ shopId: id name url city isOfficial isPowerBadge __typename }} url wishlist sourceEngine: source_engine imageUrl countReview discountPercentage rating ratingAverage labelGroups {{ position title type url __typename }} violation {{ headerText descriptionText imageURL ctaURL ctaApplink buttonText buttonType __typename }} __typename }} __typename }} __typename }}"}}]'

            # Send POST request
            for attempt in range(3):
                try:
                    response = requests.post(url_target, headers=headers, data=query)
                    response.raise_for_status()
                    break
                except requests.exceptions.HTTPError as e:
                    if response.status_code == 429:
                        print(f"Rate limit hit, waiting {delay} seconds before retry {attempt+1}/3")
                        time.sleep(delay)
                        continue
                    raise e
                except Exception as e:
                    print(f"Error on attempt {attempt+1}: {e}")
                    time.sleep(delay)
                    continue
            else:
                print(f"Failed to fetch page {page} after 3 attempts")
                continue

            # Parse JSON response
            try:
                products = response.json()[0]['data']['ace_search_product_v4']['data']['products']
                if not products:
                    print(f"No more products found on page {page}")
                    break
                all_products.extend(products)
            except (KeyError, IndexError, ValueError) as e:
                print(f"Error parsing JSON on page {page}: {e}")
                break

            # Delay to avoid rate limiting
            time.sleep(delay)

        # Process and save data
        if all_products:
            # Extract relevant fields
            processed_data = []
            for product in all_products:
                processed_data.append({
                    'name': product.get('name', ''),
                    'price': product.get('priceStr', ''),
                    'shop_name': product.get('shop', {}).get('name', ''),
                    'shop_city': product.get('shop', {}).get('city', ''),
                    'rating': product.get('ratingAverage', ''),
                    'reviews': product.get('countReview', ''),
                    'url': product.get('url', ''),
                    'is_official': product.get('shop', {}).get('isOfficial', False)
                })

            # Save to CSV
            timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
            df = pd.DataFrame(processed_data)
            filename = f'tokopedia_products_{keyword}_{timestamp}.csv'
            df.to_csv(filename, encoding='utf-8', index=False)

            print(f"Scraped {len(all_products)} products. Data saved to {filename}")
            return df
        else:
            print("No products scraped")
            return None

    except Exception as e:
        print(f"An error occurred: {e}")
        return None

def main():
    # Example keyword
    keyword = "baju anak perempuan"

    # Scrape Tokopedia data
    df = scrape_tokopedia(keyword, max_pages=5, delay=5)

    # Display sample data
    if df is not None and not df.empty:
        print("\nSample Data (first 5 rows):")
        print(df.head())

if __name__ == "__main__":
    main()
```

# **Memvalidasi Data**

*   Memastikan data akurat, konsisten, dan sesuai aturan bisnis (contoh: umur positif, pendapatan tidak negatif).
*   Melibatkan pemeriksaan logika, rentang nilai, dan format.
*   Data yang tidak valid dihapus atau diperbaiki.

```python
# Memvalidasi data
# Cek umur tidak valid (misalnya, <18 atau >100)
invalid_age = data[(data['age'] < 18) | (data['age'] > 100)]
print("\nBaris dengan umur tidak valid:")
print(invalid_age)

# Cek pendapatan negatif
invalid_income = data[data['income'] < 0]
print("\nBaris dengan pendapatan tidak valid:")
print(invalid_income)

# Cek kategori produk yang valid
valid_categories = ['Clothes', 'Electronics', 'Books']
invalid_category = data[~data['product_category'].isin(valid_categories)]
print("\nBaris dengan kategori tidak valid:")
print(invalid_category)

# Menghapus data tidak valid
data_valid = data[
    (data['age'].between(18, 100)) &
    (data['income'] >= 0) &
    (data['product_category'].isin(valid_categories))
]

# Cek setelah validasi
print("\nData setelah validasi:")
print(data_valid.info())
```
